```python
fig = px.scatter(df, x="age", y="fare", color="class", symbol="alive")
fig.show()
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.box(df, y="total_bill")
fig.show()
```

```python
df.head()
```

```python
fig2 = px.histogram(df, x="sex", y="total_bill", color="day" , pattern_shape="smoker")

fig2.show()
```

```python
df = px.data.iris() # iris is a pandas DataFrame
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
```
